```python
sample_size = 100
interval_k = population_size//sample_size
```

```python
print("Muestra obtenida por muestreo sistematico")
display(systematic_sample)
```

```python
grouped_by_platform = data.groupby("Plataforma")
grouped_by_platform.head()
```

** Planteamiento:
En el caso de una empresa necesitar una aplicacion para su empresa (vaya la redundancia), tomara datos sobre, las diferentes plataformas donde puede dessarrollarla, su costo y su tiempo de desarrollo, entonces podra determinar economicamente  y eficazmente la plataforma mas favorable para desarrollar su aplicacion**

```python
group_analysis = grouped_by_platform[['Costo (USD)','Tiempo de desarrollo (horas)']].mean()
print("Analisis por plataforma")
display(group_analysis)
```

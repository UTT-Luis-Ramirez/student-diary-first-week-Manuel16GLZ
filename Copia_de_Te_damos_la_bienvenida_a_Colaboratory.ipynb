###Muestreo aleatorio sistematico

Seleccion sistematica de la muestra

###Definicion del intervalo K (tamaño de la poblacion / el tamaño de la muestra deseada)

```
sample_size = 100
interval_k = population_size//sample_size
```

###Seleccion de muestra sistematica

```
systematic_sample = data.iloc[::interval_k]
```

```
print ("Muestra obtenida por muestreo sistematico:")
display(systematic_sample)
```

###Agrupacion de datos y subpoblaciones

Dividir la poblacion por un factor

```
grouped_by_dev_lenguage = data.groupby("Idioma de programación")
grouped_by_dev_lenguage.head ()
```

```
group_analysis = grouped_by_dev_lenguage[['Líneas de código','Número de bugs']].mean()
print("analisis por lenguaje de progamacion")
display(group_analysis)
```
